```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import pandas as pd 
```

```python
a = "diacylglyceride"
b = "diglyceride"
c = "glydiceride"
d = "dog"
# a = "lysophosphatidic acid"
# b = "1-acyl-sn-glycerol 3-phosphate"
# c = "lysophosphatidic"
# d = "acid"
```

```python
Ratio1 = fuzz.ratio(a.lower(),b.lower())
Ratiopar1 = fuzz.partial_ratio(a.lower(),b.lower())
Ratiotoken1 = fuzz.token_sort_ratio(a,b)
Ratiotokenset1 = fuzz.token_set_ratio(a, b)
print(Ratio1)
print(Ratiopar1)
print(Ratiotoken1)
print(Ratiotokenset1)
```

```python
Ratio2 = fuzz.ratio(a.lower(),c.lower())
Ratiopar2 = fuzz.partial_ratio(a.lower(),c.lower())
Ratiotoken2 = fuzz.token_sort_ratio(a,c)
Ratiotokenset2 = fuzz.token_set_ratio(a, c)
print(Ratio2)
print(Ratiopar2)
print(Ratiotoken2)
print(Ratiotokenset2)
```
